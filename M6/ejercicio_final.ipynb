### 5. Enuncie al menos tres diferencias entre el análisis factorial y el método de componentes  principales.

#### Respuesta:

A continuación, se presentan 3 distinciones entre el análisis factorial y el análisis de componentes principales:

Propósito:
El análisis de componentes principales busca explicar la mayor parte de la variabilidad de los datos utilizando un número reducido de variables en comparación con las originales. Su objetivo es reducir la cantidad de variables y explicar la varianza total de las mismas.
En cambio, el análisis factorial tiene como objetivo identificar los constructos subyacentes de las variables. Busca descubrir los factores latentes comunes que explican la correlación entre las variables originales. En este sentido, reduce los datos a factores a los que cada variable contribuye, con el propósito de encontrar relaciones entre ellas que expliquen la varianza compartida. Se centra en explicar las covarianzas o correlaciones entre las variables.

Método de construcción:
En el análisis de componentes principales, se calculan los componentes como combinaciones lineales de las variables originales, con el objetivo de explicar la mayor parte de la varianza total.
Por otro lado, en el análisis factorial, son las variables originales las que se expresan como combinaciones lineales de los factores. Busca obtener un nuevo conjunto de variables con un número reducido que represente lo que es común entre esas variables. Se parte del supuesto de que las observaciones son resultado de una transformación lineal de factores latentes de menor dimensión, con la adición de ruido gaussiano.

Interpretación de los resultados:
Dado que el objetivo del análisis de componentes principales es explicar la varianza total, los componentes en sí no pueden interpretarse directamente en términos de las variables originales, especialmente cuando hay un gran número de variables. En cambio, se ordenan según la cantidad de varianza total que cada uno explica. Este enfoque es más adecuado cuando existe una relación lineal entre las variables.
En el caso del análisis factorial, es posible visualizar la contribución de cada variable a cada factor y, por lo tanto, identificar el constructo o dimensión que subyace a ellos. Esto permite asignar un significado a los factores extraídos, etiquetándolos de acuerdo con la interpretación teórica y analizando cómo cada variable contribuye a cada factor (carga factorial).

### 6. ¿Qué significa que el método de clusterización sea jerárquico?

#### Respuesta:
Los conglomerados jerárquicos son un enfoque de agrupamiento iterativo en el cual las observaciones individuales se agrupan de manera progresiva en conglomerados cada vez más grandes. Se inicia considerando cada observación como un conglomerado individual y, en función de las distancias o similitudes entre ellos, se van fusionando en conglomerados más grandes. Este proceso continúa hasta que finalmente queda un solo conglomerado que contiene a todas las observaciones. Para visualizar las relaciones entre los conglomerados de forma jerárquica, se utiliza un dendrograma.

El dendrograma muestra las conexiones y agrupaciones entre los conglomerados de manera gráfica. En el eje vertical se representan las distancias o similitudes entre los conglomerados, y en el eje horizontal se muestran los conglomerados y las observaciones individuales. A medida que se avanza hacia la base del dendrograma, los conglomerados se fusionan en otros cada vez más grandes, hasta llegar al conglomerado final que contiene a todas las observaciones.

Es importante destacar que también es posible realizar el análisis de conglomerados en sentido inverso. En este caso, se parte de un gran conglomerado que contiene todas las observaciones y se van dividiendo en sub-conglomerados más pequeños, hasta que cada observación se convierte en un único conglomerado. En este escenario, el dendrograma se presenta en sentido inverso, con los conglomerados más grandes en la parte superior y los más pequeños en la base.

Una ventaja de este tipo de análisis de conglomerado es que no requiere especificar de antemano la cantidad de conglomerados deseada, ya que la jerarquía permite explorar los diferentes niveles de agrupación.

### Feature Extraction: Usando los datos filtrados reduzca el número de variables. Busque una representación que le permita explicar un 90% de la varianza original de los datos.

```python
# Se deja en el DataFrame solo con las columnas a utilizar en el análisis, es decir,se saca ID que es una variable categórica
# con datos distintos para cada caso, que no generan patrones y su unico sentido es identificar cada caso.
df_final = df_limpio.drop('ID',axis=1)
df_final.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Reducción de número de variables con PCA 

# Se estandarizan los datos
scaler = StandardScaler()
datos_estandarizados = scaler.fit_transform(df_final)

# Se genera y ajusta modelo PCA explicando el 90% de varianza
PCA = PCA(n_components=0.9,svd_solver = 'full')
model_PCA = PCA.fit_transform(datos_estandarizados) 
print("Cantidad de casos, y cantidad de variables: ", model_PCA.shape,"\n")

# Se observa el % de varianza explicada explicada por cada componente
explained_variance_ratio = PCA.explained_variance_ratio_
cumulative_explained_variance_ratio = np.cumsum(explained_variance_ratio)
for i, ratio in enumerate(cumulative_explained_variance_ratio):
    print('Varianza explicada por los primeros {} componentes: {:.2%}'.format(i + 1, ratio))

# Se genera un dataframe con los PCA indicados por el modelo    
df_PCA = pd.DataFrame(data=model_PCA, columns=['PC{}'.format(i+1) for i in range(model_PCA.shape[1])])
df_PCA.head()
```

```python
# Son 10 componentes lo que logran explicar el 90,1% de la varianza
```

### Clustering:

#### a) Usando los resultados anteriores clusterice el comportamiento de sus clientes. El área de  marketing solicitó que el número de clusters sea no mayor a 8. Encuentre el número óptimo de  clusters considerando esa restricción

```python
# Rango de valores de k a probar
k_values = range(1, 9)

# Lista para almacenar las puntuaciones de inercia
inertia_scores = []

# Realizar K-means para cada valor de k y almacenar la puntuación de inercia
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=15).fit(df_PCA)
    inertia_scores.append(kmeans.inertia_)

# Gráfico del método del codo
plt.plot(k_values, inertia_scores, 'bx-')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Inercia')
plt.title('Método del Codo')
plt.show()
```

```python
# El gráfico del codo muestra 3 clúster, ya que desde ahí tener un clúster más no disminuye significativamente la inercia.
```

```python
# Se observa también el gráfico de la sirueta:

k_values = range(2, 9) # Rango de valores de k a probar
silhouette_scores = [] # Lista para almacenar las puntuaciones de silueta

# Realizar K-means para cada valor de k y almacenar la puntuación de silueta
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=15).fit(df_PCA)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(df_PCA, labels))

# Gráfico del método de la silueta
plt.plot(k_values, silhouette_scores, 'bx-')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Puntuación de Silueta')
plt.title('Método de la Silueta')
plt.show()
```

```python
# Si bien los 2 clúster tienen la mayor puntuación en el gráfico de la sirueta, le sigue con un resultado alto los 3 clúster
# y como se desean utilizar los conglomerados para campañas de marketing, se utilizan los 3 del gráfico del codo. 
```
