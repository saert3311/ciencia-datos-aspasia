Con estos archivos realice los siguientes ejercicios:
1. Cargue los tres datasets usando pandas.
2. Revise los datos cargados y realice las modificaciones necesarias
para empezar a trabajar
3. Elimine las columnas que no aportan información

```python
import pandas as pd
#punto 1
state_abbrevs_df = pd.read_csv('state-abbrevs.csv', sep=';', )
state_area_df = pd.read_csv('state-area.csv', sep=';', )
state_population_df = pd.read_csv('state-population.csv', sep=';', )
#punto 2 y 3 
#limpiamos columnas que no necesitamos
state_abbrevs_df.drop(state_abbrevs_df.columns[[2,3,4,5,6]], axis=1, inplace=True)
state_area_df.drop(state_area_df.columns[[2,3,4,5,6]], axis=1, inplace=True)
state_population_df.drop(state_population_df.columns[[4,5,6]], axis=1, inplace=True)
#renombramos la columna donde estan los estados abreviados nos servira para el merge
state_abbrevs_df.rename(columns={'abbreviation':'state_short'}, inplace=True)
state_population_df.rename(columns={'state/region': 'state_short'}, inplace=True)
```

4. Agrupe por state/region y ages, para luego agregar usando
promedio, mínimo, máximo y contar

```python
agrupado_state_ages_df = state_population_df.groupby(['state_short', 'ages']).aggregate({'population':['min','max','mean','count']})
# cambiamos el tipo de dato para que sea mas leible
agrupado_state_ages_df['population'] = agrupado_state_ages_df['population'].astype('int64')
agrupado_state_ages_df
```

5. Forme un nuevo dataset llamado df que tenga la información de los
3 sets importados. Use las llaves necesarias.

```python
#previamente renombramos las columnas asi el merge automaticamente las detecta y agrupa
df = pd.merge(pd.merge(state_area_df, state_abbrevs_df), state_population_df)
df
```

6. Calcule la población por área para cada estado. Para esto agrupe por
state/region y calcule el promedio de la población. Guarde dicha
información en un data frame con las columnas 'state' y 'pop/area'
(hint: puede extraer los valores de un objeto groupby, finalizando su
sentencia con .values, o extraer sus etiquetas con .index)

```python
grouped_data = df.groupby(['state']).aggregate({'population':'mean'})
mean_population = grouped_data.iloc[:, [0]]
mean_population.reset_index()
mean_population.rename(columns={'population':'pop/area'}, inplace=True)
mean_population['pop/area'] = mean_population['pop/area'].astype('int64')
mean_population
```

7. Finalmente, cree un dataframe que posea: El estado, el área, la
población, la población por área y como índice la abreviación del
nombre del estado.
