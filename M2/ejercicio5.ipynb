Una Tabla de BINGO

Para una tarde de bingo se usarán tablas generadas con las siguientes
reglas:
● Los 5 números en cada columna son aleatorios ordenados no
repetidos de acuerdo al rango asignado por columna.
● La casilla central de la tabla tiene valor marcado cero o ‘libre’.
Realice un programa para generar una tabla de bingo.

B = entre 1 - 15
I = entre 16 - 30
N = entre 31 - 45
G = entre 46 - 50
O = entre 61 - 75

```python
import random
import pandas as pd

def generar_numeros(inicio: int, final: int, libre=False) -> list:
    if final < inicio:
        raise ValueError('El valor del final no puede ser mayor que en de inicio')

    columna = set()
    while len(columna) < 5:
        numero_generado = random.randint(inicio, final)
        columna.add(numero_generado)
    prelista = list(columna)
    if libre == True:
        prelista[2] = 'libre'
    return prelista

def generar_bingo():
    nuevo_bingo = {'B':[],'I':[], 'N':[], 'G':[], 'O':[]}
    for letra in nuevo_bingo.keys():
        match letra:
            case 'B':
                nuevo_bingo['B'].extend(generar_numeros(1,15))
            case 'I':
                nuevo_bingo['I'].extend(generar_numeros(16,30))
            case 'N':
                nuevo_bingo['N'].extend(generar_numeros(31,45, libre=True))
            case 'G':
                nuevo_bingo['G'].extend(generar_numeros(46,50))
            case 'O':
                nuevo_bingo['O'].extend(generar_numeros(61,75))
    return nuevo_bingo


nuevo_carton = pd.DataFrame(generar_bingo())
nuevo_carton
```
